```python
import seaborn as sns
```

### Categorical plots
#### To do the analysis on Categorical Data Points we can use
#### Box plot
#### Violin plot
#### Count plot
#### Bar plot

```python
df = sns.load_dataset("tips")
```

```python
## Countplot
sns.countplot('sex' , data = df) # Gives us the count of observations in each categorical bin
```

```python
sns.countplot('smoker' , data = df)
```

```python
sns.countplot( y = 'day' , data = df)
```

```python
## Barplot - Here we have to give both x value and y value
sns.barplot( x = 'total_bill' , y = 'smoker' , data = df)
```

```python
sns.barplot( x ='sex' , y = 'total_bill' , data = df)
```

```python
## Boxplot
sns.boxplot('sex' , 'total_bill' , data = df)
```

```python
sns.boxplot('day' , 'total_bill' , data = df , palette='rainbow')
# Boxplot is based on 5 number summary ( minimum , first quartile , median , third quartile , maximum ). Interquantile range = Q3-Q1.
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

```python
## Violinplot - We can see the distribution of data in terms of KDE and the boxplot
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
