```python
## Plotting using matplotlib

plt.scatter(x,y, c ='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D-graph')
plt.savefig('01_Test.png')
```

```python
# Creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'y-')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
```

```python
x = np.arange(1,20,2)
y = x**2 + 3*x + 5
plt.plot(x,y,'g*-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Quadratic graph")
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
```

```python
# Bar plots
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 

plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
    
plt.show
```

```python
# Histograms
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a , bins = 25 ) # By default bins value in the histogram will be 10
plt.title("histogram") 
```

```python
# Box plot using matplotlib
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False); 
```

```python
# Pie Charts
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0.3, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
